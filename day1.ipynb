```python
 def romanToInt( s: str) -> int:
        converter = {
            "I" : 1,
            "V" : 5,
            "X": 10,
            "L": 50,
            "C": 100,
            "D": 500,
            "M": 1000
            }
        total = 0
        l = list(s)
        for i in range(len(l)):
            # 4 and 9 check
            letter = l[i]
            if letter == "I" and i + 1 < len(l) and l[i + 1] in ["V", "X"]:
                total -= converter[letter]
            elif letter == "X" and i + 1 < len(l) and l[i + 1] in ["L", "C"]:
                total -= converter[letter]
            elif letter == "C" and i + 1 < len(l) and l[i + 1] in ["D", "M"]:
                total -= converter[letter]
            else:
                total += converter[letter]
            print(total)
        return total
```

```python
s = "MCMXCIV"
print(romanToInt(s))
```
